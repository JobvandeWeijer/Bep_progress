#### Dynamic link and added path

A link was made from the missing library libtiff.so.4 to the newer library libtiff.so.5 which on itself is a link to libtiff.so.5.5.0, after this link was added the camera software still showed that it was missing library libtiff.so.4, I searched a bit if the link was done correctly and tried to link it directly to 5.5.0 but nothing helped. Afterwards by the lnn command it was found that the path for libtiff.so.4 was not found by the camera software. 

First I tried adding the path in the .conf files but this didn't work. 

Later I used the enviromential variable LD_LIBRARY_PATH with use of a youtube video to assign the path.
