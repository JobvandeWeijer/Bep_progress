## Installation of the camera

To install the camera, a firepackage needed to be installed. This is done using a executable that can be found on the AVT  site (https://www.alliedvision.com/en/support/software-downloads/). However, this firepackage installer was made for windows. In windows it worked fine directly and it was able to get the images from the camera. Ubuntu wasn't able to execute the file at first. The windows wine application was needed to execute the file in ubuntu. After the firepackage was installed there was still a problem regarding windows and ubuntu drivers:

This showed that there was no 1394 card installed. I've searched on the AVT site and other forums how to install the linux driver/card that is needed to run the programm however I've not yet succeeded to get the good driver. When looking at the AVT site, the drivers that are made on there, are windows drivers or they are  for another type of camera (which doesn't work when installed). On forums its also hard to find the correct 1394 drivers. I've tried other drivers for different cameras of the AVT company but nothing seems to work.
