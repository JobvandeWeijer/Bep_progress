## Windows connection

For the connection of the epuck, the epic matlab tool is used.(http://www.e-puck.org/index.php?option=com_content&view=article&id=29&Itemid=27) First after installing it was not established to get a connection. After installing a bootloader this problem was solved. The open documentation of this epic tool was used to control the epuck. First there was no movement after setting the motor speeds, however it was found that the code constantly needed to be updated to change the motor settings and control the epuck. It can also be done via the joystick window in the epic but this is manually. . In the later chapters of the open documentation it was found that a controller can be send to the epuck and that it doesn't needed to be updated manually. 

The open documentation stated: "It is not possible to connect to more than one e-puck at a time due to Matlab limitations." With some trying this is possible, however it is not yet succeeded to move multiple epucks at once. Therefore, I'm not quite sure if it is possible for multiple epucks to follow each other.
